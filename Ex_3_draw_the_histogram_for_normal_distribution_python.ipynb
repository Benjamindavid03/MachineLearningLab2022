# Ex 3 : Draw the histogram for normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

N = 500000
x = np.random.randn(N) #Return a sample (or samples) from the “standard normal” distribution.
#x = stats.norm.rvs(size=N) #A normal continuous random variable.

num_bins = 20 # separators
plt.hist(x, bins=num_bins, facecolor='blue', alpha=0.5) # histogram plots

y = np.linspace(-4, 4, 1000) # Return evenly spaced numbers over a specified interval
bin_width = (x.max() - x.min()) / num_bins
plt.plot(y, stats.norm.pdf(y) * N * bin_width) #plot the line for pdf(x),y

plt.show()
```
